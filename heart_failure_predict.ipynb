Importing libs and read the dataset for first view;

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score, mean_squared_error, roc_curve, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

db = pd.read_csv(r"C:\Users\Bruno Birello\Downloads\Datasets\heart_failure.csv\heart_failure_clinical_records_dataset.csv")
db.info()
db.head(10)
```

After reading the dataset, we need to see the data, columns, how much strings and what type of info those data are

Spliting our dataset in x and y, where x represents all the columns unless the last, and y the opossite, it represents the last column that we will use to fit comparing to the x data

After Spliting our data, we need to create our model and fit it to the relevants data gathered on the dataset

```python
x = db.iloc[:, :-1]
y = db.iloc[:, -1]

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state= 10 )
model = LogisticRegression()
model_fitted = model.fit(x_train, y_train)
```

We should predict our data test that we split before for better accuracy

```python
y_pred = model_fitted.predict(x_test) 
print(accuracy_score(y_pred, y_test))

train_accuracy = model_fitted.score(x_train, y_train)
test_accuracy = model_fitted.score(x_test, y_test)

print("Train accuracy:", model_fitted.score(x_train, y_train))
print("Test accuracy:", model_fitted.score(x_test, y_test))
```

Finishing our fit-predict job, we better test and see the results by ourself, so i created a new line that will choose a random number with the gaps from the Minimum to Maximus in every columns to new predict samples.

```python
import numpy as np

min_age, max_age = min(x['age']), max(x['age'])
min_anaemia, max_anaemia = min(x['anaemia']), max(x['anaemia'])
min_creatinine, max_creatinine = min(x['creatinine_phosphokinase']), max(x['creatinine_phosphokinase'])
min_diabetes, max_diabetes = min(x['diabetes']), max(x['diabetes'])
min_ejection, max_ejection = min(x['ejection_fraction']), max(x['ejection_fraction'])
min_pressure, max_pressure = min(x['high_blood_pressure']), max(x['high_blood_pressure'])
min_platelets, max_platelets = min(x['platelets']), max(x['platelets'])
min_secreatinine, max_secreatinine = min(x['serum_creatinine']), max(x['serum_creatinine'])
min_sesodium, max_sesodium = min(x['serum_sodium']), max(x['serum_sodium'])
min_time, max_time = min(x['time']), max(x['time'])

new_line = pd.DataFrame({
    'age': np.random.randint(min_age, max_age, size=1),
    'anaemia': np.random.randint(min_anaemia, max_anaemia, size=1),
    'creatinine_phosphokinase': np.random.randint(min_creatinine, max_creatinine, size=1),
    'diabetes': np.random.randint(min_diabetes, max_diabetes, size=1),
    'ejection_fraction': np.random.randint(min_ejection, max_ejection, size=1),
    'high_blood_pressure': np.random.randint(min_pressure, max_pressure, size=1),
    'platelets': np.random.randint(min_platelets, max_platelets, size=1),
    'serum_creatinine': np.random.randint(min_secreatinine, max_secreatinine, size=1),
    'serum_sodium': np.random.randint(min_sesodium, max_sesodium, size=1),
    'sex': np.random.choice([0, 1]),  
    'smoking': np.random.choice([0, 1]), 
    'time': np.random.randint(min_time, max_time, size=1)
})

previsao = model_fitted.predict(new_line)
limite = 0.5
print(new_line)

if previsao[0] > limite:
    print("There is a high possibility of you be affected a heart failure.")
else:
    print("There is a low possibility of you be affected a heart failure.")
```
